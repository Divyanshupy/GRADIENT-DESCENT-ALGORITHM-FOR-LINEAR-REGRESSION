```
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```
np.random.seed(0)
x =np.random.randn(100)
y = 2 * x + 5 + np.random.randn(100)

b = 0
w = 0
m = x.shape[0]
print(m)
alpha = 0.01
```

```
print(y.shape)
```

```
plt.scatter(x,y)

```

```
y_hat_arr = []

for i in range(110):
    y_hat = w * x + b 
  
    dw = (-2/m)*sum(x*(y-y_hat))
    db = (-2/m)*sum(1*(y-y_hat))
    w =  w - alpha * dw
    b =  b - alpha *db
    y_hat_arr.append(y_hat)
    
    plt.scatter(x,y)
    plt.plot(x,y_hat,'red')
    plt.savefig('work{}'.format(i))
    plt.title("m = {0:0.2f} ,c = {1:0.2f}".format(w,b))
    
    plt.show()
  
    
    
print(w,b)
```
